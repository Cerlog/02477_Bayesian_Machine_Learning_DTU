# Part 1: Gaussian Processes and Covariance Functions

In this part, we will study covariance functions for Gaussian process models. Consider the following six covariance functions:

$$
k_1(x, x') = 2 \exp\left(\frac{-(x - x')^2}{2 \cdot 0.3^2} \right) \quad \quad (1)
$$

$$
k_2(x, x') = \exp\left(\frac{-(x - x')^2}{2 \cdot 0.1^2} \right) \quad \quad (2)
$$

$$
k_3(x, x') = 4 + 2xx' \quad \quad (3)
$$

$$
k_4(x, x') = \exp\left(-2 \sin(3\pi \cdot |x - x'|)^2 \right) \quad \quad (4)
$$

$$
k_5(x, x') = \exp\left(-2 \sin(3\pi \cdot |x - x'|)^2 \right) + 4xx' \quad \quad (5)
$$

$$
k_6(x, x') = \frac{1}{5} + \min(x, x') \quad \quad (6)
$$

Some of them should be familiar, and some of them might be new to you.

### Task 1.1: Determine the analytical marginal prior mean and variance of a Gaussian process, 

$$ f_i(x) \sim GP(0, k_i(x, x')) $$ 

i.e., compute $\mathbb{E}[f_i(x)]$ and $\mathbb{V}[f_i(x)]$ for each of the six covariance functions (for $i = 1,2,\dots,6$).


Since, the Gaussian Processes are defined with $f(x) \sim \mathcal{N}(0, k(x, x'))$, the mean for all functions is 0. The variance is calculated as $\mathbb{V}[f_i(x)] = Cov(f_i(x), f_i(x)) = k_i(x, x)$, therefore we just need to set $x = x'$ in the kernel functions, to obatain the variance of the Gusssian Process.

### Task 1.2: Which of the six covariance functions are stationary covariance functions?

Given in the notebook from week 5, we know that the **stationary** kernel means that the covariance function only depends on the difference between two points, i.e. $\tau = \mathbf{x}_n - \mathbf{x}_m$ for two points in the input space $\mathbf{x}_n, \mathbf{x}_m \in \mathbb{R}^D$:

We, observe, that the kernel functions:
- $k_1(x, x')$ depends on the distance between $x$ and $x'$, therefore it is a stationary kernel. 
- $k_2(x, x')$ is also stationary kernel, as it is same as $k_1(x, x')$, only scaled by 2. 
- $k_3(x, x')$ is not stationary kernel, as it depends on the product of $x$ and $x'$.
- $k_4(x, x')$ is stationary kernel, as it depends on the distance between $x$ and $x'$.
- $k_5(x, x')$ is similar to $k_4(x, x')$, however it also depends on the product of $x$ and $x'$, therefore it is not a stationary kernel.
- $k_6(x, x')$ is not a stationary kernel, as minimum does not take the difference between $x$ and $x'$.
